# Quantum Oscillators #

Our classical simulations have largely centered on either harmonic oscillators, where the potential is of the form
$$V(x) = \frac{m \omega_o^2 x^2}{2} $$
or Morse oscillators whose more complicated potential resembles the harmonic form at low amplitudes but becomes increasingly anharmonic at larger displacements. We saw that the harmonic oscillator provides a simple illustration for linear processes like absorption, while the anharmonic form of the Morse oscillator was required to capture nonlinear signals like pump-probe and 2D spectroscopies. 

In this exercise, we'll examine the quantum mechanical form of oscillators like these and see where they are both similar and different from the classical results. 

## Quantum Harmonic Oscillator ##

The starting point for our discussion is the quantum harmonic oscillator -- one of the few physically relevant potentials in physics whose properties can be completely evaluated analytically. The quantum Hamiltonian for this system is
$$ H = \frac{m \omega_o^2 x^2}{2} + \frac{p^2}{2m} .$$
The operators $x$ and $p$ represent position and momentum, respectively. In the position basis, the $x$ operator corresponds simply to multiplication by $x$, and the $p$ operator takes the form
$$ p = -i \hbar \frac{d}{dx} .$$
Our first objective is to find the eigenfunctions (and associated eigenvalues) of this Hamiltonian. 

## Raising and Lowering Operators ##

The simplest route to solving this problem is to notice that the operator 
$$ a = \sqrt{\frac{m\omega_o}{2\hbar}} \left( x + \frac{i p}{m \omega_o} \right) $$
and its Hermitian conjugate
$$ a^\dagger = \sqrt{\frac{m\omega_o}{2\hbar}} \left( x - \frac{i p}{m \omega_o} \right) $$
have rather unique properties. 

## Commutator of $a$ and $a^\dagger$ ##
Note first their commutator:
$$ \left [ a, a^\dagger \right] = \frac{m\omega_o}{2\hbar} \cdot \frac{1}{m \omega_o} \left( \left [ x, - ip \right] + \left [ ip, x \right] \right) = \frac{-2i}{2\hbar} \left [ x, p \right] = 1 ,$$
where in the last step we have used the canonical commutation relation
$$ \left [ x, p \right] = i \hbar .$$

##  $H$ in terms of $a$ and $a^\dagger$ ##

With this definition, it is straightforward to check that the Hamiltonian can be expressed simply as 
$$ H = \hbar \omega_o \left( a^\dagger a + \frac{1}{2} \right) .$$

## Commutator with $H$ ##

Because $H$ can be expressed in terms of $a$ and $a^\dagger$, their commutators with $H$ are easy to compute:
$$ [H, a] = \hbar \omega_o [ a^\dagger a, a ] = \hbar \omega_o \left( a^\dagger a a - a a^\dagger a \right) = \hbar \omega_o \left( a^\dagger a - a a^\dagger \right) a = - \hbar\omega_o a ,$$
and
$$ [H, a^\dagger] = \hbar \omega_o [ a^\dagger a, a^\dagger ] = \hbar \omega_o \left( a^\dagger a a^\dagger - a^\dagger a^\dagger a \right) = \hbar \omega_o a^\dagger \left( a a^\dagger - a^\dagger a \right) = \hbar \omega_o a^\dagger ,$$


## $a$ as a "lowering" operator ##

With these commutation relations in hand, its easy to "discover" why these two operators are so interesting. Suppose that $\psi$ is an eigenstate of $H$ with eigenvalue $\lambda$. Then
$$ H a \psi = \left( a H + [H, a]\right) \psi = \left( \lambda - \hbar \omega_o \right) a \psi .$$
Thus, $a \psi$ is *also* an eigenfunction of $H$ with eigenvalue $\lambda - \hbar \omega_o$, lower by the "step size" $\hbar \omega_o$ from the original eigenfunction $\lambda$. 

For this reason, $a$ is often called the "lowering" operator. If we can find a single eigenfunction $\psi$, this means we can construct a "ladder" of states going down in energy from the original eigenvalue $\lambda$. (We'll see in a moment that whether or not that ladder has a bottom rung depends on the value of $\lambda$.)


## $a^\dagger$ as a "raising" operator ##

A similar relationship holds for $a^\dagger$. If $\psi$ is an eigenfunction of $H$ with eigenvalue $\lambda$, then
$$ H a^\dagger \psi = \left( a^\dagger H + [H, a^\dagger]\right) \psi = \left( \lambda + \hbar \omega_o \right) a^\dagger \psi .$$
so $a^\dagger \psi$ is an eigenfunction of $H$ with eigenvalue $\lambda + \hbar \omega_o$. 

For this reason $a^\dagger$ is often called a "raising" operator, since it produces a "ladder" of eigenvectors going *up*, in contrast to the "down" ladder that $a$ builds. 


## The ladder has a lower bound ##

From what we've said so far, it isn't obvious whether there are upper or lower bounds on the eigenvalues of $H$. Can the ladder of states go up and down infinitely many times? Or are there upper and lower limits, beyond which there are no more states? 

We can answer these questions by checking the normalization of the new states produced by $a$ and $a^\dagger$. Let us assume that the original eigenstate $\psi$ is normalized so that 
$$ \| \psi \|^2 = \left \langle \psi | \psi \right \rangle = 1 .$$
Then
$$ \| a \psi \|^2 = \left \langle \psi | a^\dagger a | \psi \right \rangle = \left \langle \psi \left | \left( \frac{H}{\hbar \omega_o} - \frac{1}{2} \right) \right | \psi \right \rangle = \frac{\lambda}{\hbar\omega_o } - \frac{1}{2} . $$

There should be something unnerving about this result. The norm of a function *by construction* must always be nonnegative, so it *must* be the case that 
$$ \| a \psi \|^2 \geq 0 . $$
But if it should happen to be true that $\frac{\lambda}{\hbar \omega_o} < \frac{1}{2}$, then this condition would be violated. 

The only mathematically consistent conclusion is that *for any valid eigenvalue $\lambda$*:
$$ \lambda \geq \frac{\hbar \omega_o}{2} . $$


Thus there *is* a lower bound on the spectrum of possible energy states: the ladder cannot go down forever. In contrast, the normalization condition produces no such restriction on going *up* the ladder. We can apply $a^\dagger$ as many times as we like, and we'll always obtain a valid eigenfunction. 

## Allowed values of $\lambda$ ##

Finally, we're ready to "unmask" the true value of $\lambda$. In fact, this emerges naturally by resolving the apparent conflict between the lower-bound condition we just derived and the fact that applying $a$ should always result in an eigenstate of lower energy. Recall that so far we've derived two central results:

$$ H a \psi = \left( \lambda - \hbar \omega_o \right) a \psi$$

and 

$$ \| a\psi \|^2 = \frac{\lambda}{\hbar\omega_o } - \frac{1}{2} . $$

How can these statements both be true in general? The first one implies that, from any initial eigenvalue $\lambda$, $n$ successive applications of $a$ will produce an eigenvector with eigenvalue $\lambda - n \hbar \omega_o$. But for any finite value of $\lambda$, this means that there will always be *some* value of $n$ such that 
$$ \lambda - n \hbar \omega_o < \frac{\hbar \omega_o}{2} . $$
Thus it appears that we must ultimately produce a state with energy lower than our lower bound of $\frac{\hbar\omega_o}{2}$, violating the (foundational) condition that vector norms must always be positive. 

Fortunately, there is an escape from this contradiction: *if $a \psi$ is identically zero*, then the equation 
$$ H a \psi = \left( \lambda - \hbar \omega_o \right) a \psi$$
is satisfied automatically, regardless of the values of $\lambda$, $\omega_o$, and even $\hbar$. Thus, as we move down our ladder of states, if we eventually reach a state where $a \psi = 0$, then our ladder naturally terminates; the "lowering" equation is satisfied without depending on the existence of another state with lower energy. 

Now, if $a \psi = 0$ identically, then also $\| a \psi\|^2 = 0$, which (from our normalization condition) implies that 
$$ 0 = \| a\psi \|^2 = \frac{\lambda}{\hbar\omega_o } - \frac{1}{2} = 0 $$
or simply
$$\lambda = \frac{\hbar \omega_o}{2} . $$

Thus *the smallest allowed eigenvalue* for $H$ is 
$$ \lambda_0 = \frac{\hbar \omega_o}{2} ,$$
and all higher-energy states must have energies
$$ \lambda_n = \frac{\hbar \omega_o}{2} \left( n + \frac{1}{2} \right) . $$

## Normalization of eigenstates ##

What's left at this point is just house-keeping. Let us denote by $\psi_n$ the *normalized* eigenvector of $H$ whose eigenvalue is $\lambda_n$. Applying $a$ to the $n^\text{th}$ eigenstate $\psi_n$, we obtain for the norm of the new state
$$ \| a\psi_n \|^2 = \left \langle \psi_n \left | \left( \frac{H}{\hbar \omega_o} - \frac{1}{2} \right) \right | \psi_n \right \rangle = \frac{\hbar \omega_o \left( n + \frac{1}{2} \right)}{\hbar\omega_o } - \frac{1}{2} = n . $$
Since we have set by convention $\| \psi_{n-1} \|^2 = 1$, this implies that
$$ a \psi_n = \sqrt{n} \psi_{n-1} . $$
A similar calculation for $a^\dagger \psi_n$ shows that
$$ a^\dagger \psi_n = \sqrt{n+1} \psi_{n+1} . $$
Thus the $n^\text{th}$ eigenstate can be expressed concisely as 

$$ \psi_n = \frac{a^n}{\sqrt{n!}} \psi_o .$$


## Position representation of the ground state ##

Finally, let's calculate the form for the "ground state" $\psi_o$ in the position representation. To do this, we start with the defining condition that
$$ a \psi_o = 0, $$
which is unique to the ground state. In terms of the position representation of the $x$ and $p$ operators, this becomes
$$ a \psi_o = \sqrt{\frac{m\omega_o}{2\hbar}} \left( x + \frac{\hbar}{m \omega_o} \frac{d}{dx} \right) \psi_o = 0 $$
or simply 
$$ \frac{d}{dx} \psi_o = - \frac{m \omega_o}{\hbar} x \psi_o . $$
This differential equation is solved by the function 
$$\psi_o(x) = A_o e^{-\frac{m \omega_o}{2 \hbar} x^2 } , $$
where $A_o$ is a normalizing constant to ensure that $\| \psi_o \|^2 = 1$. From the Gaussian integral $\int e^{-\frac{x^2}{2\sigma^2}} dx = \sqrt{2 \pi \sigma^2}$, this condition is satisfied by setting 
$$A_o^2 = \sqrt{\frac{m\omega_o}{\pi \hbar} },$$
leading to the final form
$$ \psi_o(x) = \left( \frac{m\omega_o}{\pi \hbar} \right)^\frac{1}{4} e^{-\frac{m \omega_o}{2 \hbar} x^2 
} . $$

If we denote
$$\sigma = \sqrt{\frac{\hbar}{m\omega_o}}, $$
this becomes simply
$$ \psi_o(x) = \frac{1}{\sqrt{\pi^{1/2} \sigma}} e^{- \frac{x^2}{2 \sigma^2} } .$$


## Position representation of the excited states ##

The position representation of the excited states can now be calculated recursively from the ground state wavefunction. Using the raising operator
$$ a^\dagger = \frac{1}{\sqrt{2}} \left( \sigma^{-1} x - \frac{d}{d(\sigma^{-1}x)} \right) , $$
the excited-state wavefunctions are 
$$ \psi_n(x) = \frac{1}{\sqrt{2^n n!}} \left( \sigma^{-1} x - \frac{d}{d(\sigma^{-1}x)} \right)^n \psi_o(x).$$


It's a little tedious to calculate the wavefunctions by hand, so we'll do it automatically using the Python SymPy library. To simplify the calculations, let's first denote
$$z = \sigma^{-1} x $$
and 
$$ f_n(z) = \sqrt{2^n n!} \psi_n(\sigma z)$$
so that
$$ f_n(z) = \left( z - \frac{d}{dz} \right)^n f_o(z).$$
The code block below calculates the first `N` of these functions $f_n(z)$ using SymPy. 

```python
import sympy as sy
from IPython.display import display

Nbasis = 12

# Note: In the final step, we both multiply and divide by F0 so as to 
# force SymPy to factor out the exponential. (Otherwise, it's distributed
# term-by-term, which is a little messy.)

sy.var('z')
F0 = sy.exp(-z**2/2)
F = sy.exp(-z**2/2)
for n in range(1, Nbasis):
    F = z*F - sy.diff(F)
    display(sy.expand(F/F0)*F0)
```

Note that each term has the form of an $n^\text{th}$-order polynomial multiplied by the ground-state exponential. The polynomials here are the "Hermite polynomials" -- though you don't need to know that name to use them to solve problems. 

What we *do* need to do is to evaluate these functions numerically. To do that, we'll collect their polynomial coefficients in a list called ``HnCoeffs``, which will allow us to evaluate numerical values in NumPy. The next code block does exactly this. 

```python
import sympy as sy
from IPython.display import display

HnCoeffs = [[1]]

sy.var('z')
F0 = sy.exp(-z**2/2)
F = sy.exp(-z**2/2)
for n in range(1, Nbasis):
    F = z*F - sy.diff(F)
    
    print('********************************************')
    print("n = " + str(n) + ':')
    display(sy.expand(F/F0)*F0)
    
    # Create a polynomial object
    ply = sy.poly(sy.expand(F/F0))
    
    # Get the coefficients
    coeffs = ply.all_coeffs()
    print(coeffs)
    print('********************************************')
    
    HnCoeffs.append(coeffs)
```

Finally, we can numerically evaluate all harmonic oscillator wavefunctions using the expression 
$$ \psi_n(x) = \frac{1}{\sqrt{2^n n! \pi^{1/2} \sigma}} H_n \left(\frac{x}{\sigma} \right) e^{-\frac{x^2}{2 \sigma^2}} $$
with
$$\sigma = \sqrt{\frac{\hbar}{m\omega_o}}. $$

The code block below plots the first few eigenfunctions of the harmonic oscillator, with the potential energy surface plotted in the background for reference:


```python
import math
%matplotlib inline
import matplotlib.pyplot as plt
from IPython import display
import time
import numpy as np

hbar = 6.626e-34/(2.0*np.pi)

wo = (2.0*np.pi)*50e12      # Oscillator (angular) frequency
E0 = hbar*wo                # Energy of harmonic states

M = 12*(1.66054e-27)        # Mass in kg
sigma = np.sqrt(hbar/(M*wo))

# x axis for wavefunctions
xaxis = np.arange(-1,1,0.001)*(sigma)*10

PsiArray = np.zeros((len(xaxis),Nbasis))
# Calculate wavefunctions
for n in range(0, Nbasis):
    
    # Overall prefactor
    prefac = 1.0/np.sqrt((2**n)*(np.math.factorial(n))*np.sqrt(np.pi)*sigma)
    
    # Calculate the Hermite polynomial
    hn = np.zeros(np.shape(xaxis))
    for k in range(0, n+1):
        
        # The kth entry in the vector is the coefficient
        # for the x^(n-k) term. Note the int() call is necessary
        # because HnCoeffs[n][k] is a symbolic object. 
        hn += int(HnCoeffs[n][k])*np.power(xaxis/sigma, (n-k))
        
    # Exponential term
    PsiArray[:,n] = prefac*hn*np.exp(-xaxis**2/(2.0*sigma*sigma))
    

# Now let's plot the wavefunctions    
plt.figure(figsize=(6,8))
blue = np.array([0.5, 0.5, 1])
scfac = np.max(PsiArray[:,0])*2.5
for n in range(0, Nbasis):
    plt.fill_between(xaxis, PsiArray[:,n]/scfac + n+0.5, 0*xaxis+n+0.5, color=blue)
    plt.plot(xaxis, PsiArray[:,n]/scfac + n+0.5, color=blue*0.5)
    plt.text(-sigma*5.75, n+0.65, '$\\varepsilon_' + str(n) + ' = \\frac{' + str(2*n+1) + '}{2} \\hbar \\omega_o$', fontsize=12, ha='left')
    
# For comparison, plot the potential 
plt.plot(xaxis, 0.5*M*wo**2*xaxis**2/(hbar*wo), 'k')
plt.ylim(-0.5, Nbasis+0.5)
plt.xlim(-6*sigma, 6*sigma)
plt.show()
```

Note that as the wavefunctions get higher in energy, more and more population is concentrated at the "turning points", i.e., the extreme values of $x$ where (in classical harmonic motion) the momentum temporarily goes to zero (and thus the oscillator spends a good deal of time) as the oscillator turns around. This is an example of the *correspondence principle*: as the energy of the quantum system increases, its statistical properties begin to resemble those of the corresponding classical system, as reflected in this case by increased probability of "sitting" near the turning points. 

## Quantum Morse Oscillator ##

Let's now compare how these harmonic oscillator wavefunctions compare to the Morse potential that we've been using for nonlinear response simulations. For the first few energy levels, we expect the harmonic oscillator wavefunctions to be rather close to the true Morse wavefunctions, since near the bottom of the potential the behavior is mostly harmonic. At higher energies, however, we can expect substantial differences. 

The code block below plots the first few harmonic oscillator wavefunctions, with the *Morse* potential shown for reference. 

```python
Ndiss = 20                 # Dissociation number
De = hbar*wo*(Ndiss+0.5)    # Dissociation energy
alpha = np.sqrt((M*wo**2)/(2.0*De))

def morse_potential(X):
    return De*(1.0 - np.exp(-np.sqrt(M*wo**2/(2.0*De))*X))**2

# Plot the HO wavefunctions    
plt.figure(figsize=(6,8))
blue = np.array([0.5, 0.5, 1])
scfac = np.max(PsiArray[:,0])*2.5
for n in range(0, Nbasis):
    plt.fill_between(xaxis, PsiArray[:,n]/scfac + n+0.5, 0*xaxis+n+0.5, color=blue)
    plt.plot(xaxis, PsiArray[:,n]/scfac + n+0.5, color=blue*0.5)
    plt.text(-sigma*5, n+0.65, '$\\varepsilon_' + str(n) + ' = \\frac{' + str(2*n+1) + '}{2} \\hbar \\omega_o$', fontsize=12, ha='left')
    
# For comparison, plot the potential 
plt.plot(xaxis, morse_potential(xaxis)/(hbar*wo), 'k')
plt.ylim(-0.5, Nbasis+0.5)
plt.show()
```

As expected, the ground state wavefunction seems to "match" the potential reasonably well, since the potential here is very close to the corresponding harmonic surface. But as the energy increases, the quantum "turning points" get farther an farther from the "sides" of the potential, reflecting the fact that the harmonic wavefunctions are *not* good eigenfunctions for the Morse potential energy surface. 

Even though the harmonic wavefunctions aren't good eigenfunctions for the Morse potential, we *can* use them as an orthonormal basis to numerically *construct* eigenfunctions for the Morse potential -- or, in principle, any bounded one-dimensional potential. 

Given any Hamiltonian of the form
$$ H = \frac{p^2}{2m} + V(x), $$
we can calculate eigenfunctions by diagonalizing the Hamiltonian matrix defined by the matrix elements
$$ H_{mn} = \left \langle \psi_m \left | \frac{p^2}{2m} \right| \psi_n \right \rangle + \left \langle \psi_m \left | V(x) \right| \psi_n \right \rangle$$
in the harmonic oscillator basis. The eigenvectors of this Hamiltonian then "code" for the coefficients that must be used to re-weight the harmonic oscillator wavefunctions to construct, for example, Morse oscillator eigenfunctions. 

The $V(x)$ matrix elements are easily calculated numerically. The $p^2$ matrix elements are more efficiently calculated analytically by expressing $p$ in terms of raising and lowering operators:
$$ p = i \sqrt{\frac{\hbar m \omega_o}{2}} \left( a^\dagger - a \right) . $$
Then we can calculate directly the matrix elements for the $p$ operator:
$$ p_{mn} = \left \langle \psi_m \left | p \right| \psi_n \right \rangle = i \sqrt{\frac{\hbar m \omega_o}{2}} \left \langle \psi_m \left | \left( a^\dagger - a \right) \right| \psi_n \right \rangle = i \sqrt{\frac{\hbar m \omega_o}{2}} \left( \sqrt{n+1} \delta_{m(n+1)} - \sqrt{n} \delta_{m(n-1)} \right) . $$
The factors of $\sqrt{n}$ and $\sqrt{n+1}$ come from the application of $a$ and $a^\dagger$, while the Kronecker delta functions come from the restriction that the wavefunction indices must match *after* application of the raising/lowering operator. (Otherwise the inner product will vanish due to the orthonormality of the harmonic oscillator eigenfunctions.)

Once the $p_{mn}$ matrix has been constructed, we can calculate the $(p^2)_{mn}$ matrix simply through matrix multiplication of the $p$ matrix with itself.

```python
Pmat = np.zeros((Nbasis,Nbasis), dtype='complex')

# First we add all elements above the diagonal
for n in range(0, Nbasis-1):
    Pmat[n,n+1] = 1j*np.sqrt(0.5*hbar*M*wo)*np.sqrt(n+1)
    
# Then we get the elements below the diagonal by just adding the transpose
Pmat -= Pmat.T

P2mat = Pmat@Pmat

Kmat = P2mat/(2*M)

plt.imshow(np.abs(Kmat))
```

Once we have the potential and kinetic-energy matrices constructed, we can add them together to produce the full Hamiltonian matrix, denoted ``Ham`` in the code block below. We then call the NumPy `linalg.eigh` routine to calculate eigenvalues and eigenvector coefficients; the Morse wavefunctions are then calculated as linear combinations
$$ \phi_n = \sum_k D_{nk} \psi_k , $$
where $\psi_k$ is the $k^\text{th}$ harmonic oscillator wavefunction, $\phi_n$ is a Morse oscillator wavefunction, and $D_{nk}$ is the "eigenvector matrix" whose $n^\text{th}$ column contains the weighting coefficients needed to construct $\phi_n$ from the $\psi_k$ basis. 

```python
Vmat = np.zeros((Nbasis,Nbasis), dtype='complex')

Vx = morse_potential(xaxis)
for n in range(0, Nbasis):
    for m in range(0, Nbasis):
        Vmat[m,n] = np.trapz(PsiArray[:,m]*Vx*PsiArray[:,n], xaxis)

Ham = Vmat + Kmat

# Do diagonlization numerically
evals, evecs = np.linalg.eigh(Ham)

# evals stores the eigenvalues, i.e., the energy of each state
# evecs stores the eigenvectors, i.e., the weighting coefficients used to
#   construct the Morse eigenfunctions as a linear combination of harmonic
#   eigenfunctions. 

Nmax = 7

for n in range(0, Nmax):
    
    # Energy of the nth state
    En = evals[n]
    
    # Wavefunction for the nth state
    psi = np.zeros(np.shape(xaxis))
    for k in range(0, Nbasis):
        psi += np.real(PsiArray[:,k]*evecs[k,n])
        
    plt.fill_between(xaxis, psi/scfac + n+0.5, 0*xaxis+n+0.5, color=blue)
    plt.plot(xaxis, psi/scfac + n+0.5, color=blue*0.5)
    plt.text(-sigma*8, n+0.65, '$\\varepsilon_' + str(n) + ' \\approx ' + '{:.2f}'.format(En/(hbar*wo)) + ' \\hbar \\omega_o$', fontsize=12, ha='left')

# For comparison, plot the potential 
plt.plot(xaxis, morse_potential(xaxis)/(hbar*wo), 'k')
plt.ylim(-0.5, Nmax+0.5)
plt.show()
```
